Here, we use Keras with tensorflow backend to predict the use of TelCom in Milano, one timesteps ahead, using the six last timesteps of our data. 

We choose to use a RNN, more specifically a LSTM network.


```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
from src.preprocess_tools import *
import os
import math
from sklearn.preprocessing import MinMaxScaler, FunctionTransformer,StandardScaler
np.random.seed(1)
```

Transform functions

```python
def cycle_theta(t,freq):
    return(2*math.pi*(t/float(freq-1)))
def cycle_of_time(t, freq,func = math.cos):
    return func(cycle_theta(t,freq))
def isweekend(Wday):
    if Wday<=4:
        return 0
    else:
        return 1
```

Data import

```python
file = 'data/MI_squares/5561/5561_all.csv'
# dataset = pd.read_csv(file, sep = '\t', nrows=1000) #NROWS !!
dataset = pd.read_csv(file, sep = '\t')
dataset.drop("Square",axis=1,inplace=True)
dataset.set_index("Time",inplace=True)

dataset["Monthcos"] = dataset["Month"].apply(lambda x : cycle_of_time(x,12))
dataset["Monthsin"] = dataset["Month"].apply(lambda x : cycle_of_time(x,12,func = math.sin))
dataset["Hourcos"] = dataset["Hour"].apply(lambda x : cycle_of_time(x,24))
dataset["Hoursin"] = dataset["Hour"].apply(lambda x : cycle_of_time(x,24,func = math.sin))
dataset["WEnd"] = dataset["WDay"].apply(isweekend)
dataset["WDaycos"] = dataset["WDay"].apply(lambda x : cycle_of_time(x,7))
dataset["WDaysin"] = dataset["WDay"].apply(lambda x : cycle_of_time(x,7,func=math.sin))
dataset = dataset.drop(["Month","Hour","WDay","Day"],axis=1)
dataset.head(7)
```

How long do we look back in time ? (U =  10 minutes)

```python
past_n = 6
# past_n = 6*3
```

Smoothing the data ?

```python
smooth = "smooth7" # set to "" for no smoothing
win_size = 7
scaler_name = ""
if smooth:
    dataset.values[1:-1,:] = pd.rolling_mean(dataset, window = win_size, center=True).values[1:-1,:]
    scaler_name = "smooth_"
```

```python
X = np.array([scaled_data[i-past_n:i,:] for i in np.arange(past_n,scaled_data.shape[0])])
Y = dataset.values[past_n:,:5]
```

```python
X.shape, Y.shape, X[-1], Y[-1] #ok
```

#### Data spitting
Keeping the last 25% imestamps as we work with time series

```python
n_ten_min = int(dataset.shape[0]*0.75) #75% vs 25% 
X_train, Y_train = X[:n_ten_min, :,:], Y[:n_ten_min,:]
X_test, Y_test = X[n_ten_min:, :,:], Y[n_ten_min:,:]
X_train.shape, Y_train.shape, X_test.shape, Y_test.shape
```

```python
from sklearn.metrics import mean_squared_error
from keras.models import Sequential
from keras.layers import Dense, Dropout, LSTM,BatchNormalization
from keras.initializers import RandomNormal
from keras.callbacks import ModelCheckpoint
```

```python
def build_lstm_model(dpt = 0.3, loss = "mae"): 
    model = Sequential()
    model.add(LSTM(45,input_shape = X_train[0].shape, dropout = dpt, return_sequences=True))
    model.add(BatchNormalization())
    model.add(LSTM(45, input_shape=X_train[0].shape, dropout = dpt, return_sequences=False))
    model.add(BatchNormalization())
    model.add(Dropout(dpt))
    model.add(Dense(30, activation = "relu"))
    model.add(BatchNormalization())
    model.add(Dropout(dpt))
    model.add(Dense(5, activation = "relu"))
    model.compile(loss=loss, optimizer='adam')
    return model
```

```python
def fit_and_show(dpt = 0.4, loss = "mae"):
    best_w_path = 'tmp/weights.hdf5'
    m1 = build_lstm_model(dpt= dpt, loss=loss)
#     checkpointer = ModelCheckpoint(filepath=best_w_path, verbose=1, monitor='val_loss', save_best_only=True)
#     history = m1.fit(X_train, Y_train, epochs=8, batch_size=80, validation_data=(X_test, Y_test), verbose=0, shuffle=False, callbacks = [checkpointer])
    history = m1.fit(X_train, Y_train, epochs=20, batch_size=80, validation_data=(X_test, Y_test), verbose=1, shuffle=False)
#     m1.load_weights(best_w_path)
    
    # plot history
    plt.plot(history.history['loss'], label='train')
    plt.plot(history.history['val_loss'], label='test')
    plt.legend()
    plt.title("Losses - lowest val loss : "+ str(min(history.history['val_loss'])))
    plt.savefig("ML_report/5561_"+scaler_name+"_losses.png", transparent=True)
    plt.show()
    return m1, history
m1, histo1 = fit_and_show(dpt = 0.3, loss = "mae")
```

#### WORK IN PROGRESS Prédiction point par point
Ici nous tentons de prévoir les résultats plusieurs timesteps à l'avance. Keras ne semble pas suivre ces instructions...

```python
# n = 6*12
n=6
def predict_n_next(model, n, where = 0, verbose = False):
    current_past = np.expand_dims(X_test[where,:,:],axis=0).copy()
    test = np.expand_dims(X_test[where,:,:],axis=0)
    n_next = []
    for i in range(1,n+1):
        prediction = model.predict(current_past)+0.001
        n_next.append(prediction)
        if verbose : print(prediction)
        current_past = np.expand_dims(X_test[where+i,:,:],axis=0).copy()
        if verbose : print(current_past[0,:,:5])
        new_features = scaler5.transform(prediction)
        if verbose : print(new_features)
        current_past[0,-1,:5] = new_features.copy()
        if verbose : print(current_past[0,:,:5], "        \n")
    return n_next

ts = []
pap = []
# p = predict_n_next(m1,n, where = 0)
# pap.append(p)
W = np.arange(0, X_test.shape[0]-n-1, n+12)
for w in W:
    p = predict_n_next(m1,n, where = w, verbose=False)
    ts+=list(range(w,w+n))
    pap.append(p)
pap = np.squeeze(np.concatenate(pap))
print(pap.shape)
ts = np.array(ts)
```

```python
start = n_ten_min-150
for i, name in enumerate(target_names) :
    plt.figure(figsize = (10,5))
#     plt.figure(figsize = (20,10))
    plt.plot(np.arange(start,Y_train.shape[0]),Y_train[start:,i], label = "Training data")
    plt.plot(Y_train.shape[0]+ts,pap[:,i], label = "Forecasted")
    plt.plot(np.arange(Y_train.shape[0],Y_train.shape[0]+Y_test.shape[0]),Y_test[:,i], label = "Ground truth")
    plt.xticks(Y_train.shape[0]+W)
    plt.title("Validation data forecast vs ground-truth for " + name)
    plt.legend()
```
