```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
from src.preprocess_tools import *
import os
```

```python
file = 'data/MItoMI-2013-12-01.txt'
verbose = True
d = pd.read_csv(file, sep = '\t',header=None)
if verbose:
    print("Read : " +file)
d.columns = ['Time','Id1','Id2','Signal']
```

- filter and fill missing values. We keep a close square around the city, defined by :

x : 32 à 62 (étendue : 31)

y : 20 à 58 (étendue 39)

```python
mask = np.logical_and(d['Id1'].apply(lambda x : keep_a_square(x,xmin=32,xmax=62,ymin=20,ymax=58)),d['Id2'].apply(lambda x : keep_a_square(x,xmin=32,xmax=62,ymin=20,ymax=58)))
if verbose:
    print("Kept standard squares.")
d = d[mask]
d.fillna(0)
d.head(5)
```

```python
dPow = d
dPow = dPow.set_index(["Id1", "Id2"])
dPow = d.groupby(["Id1", "Id2"]).sum()
dPow.reset_index(["Id1", "Id2"],inplace=True)
```

```python
dPow = dPow.drop("Time",axis=1)
dPow.head(5)
```

```python
idx1 = np.unique(dPow["Id1"].data)
idx2 = np.unique(dPow["Id2"].data)
idx = np.unique(np.array(idx1.tolist() + idx2.tolist()))
```

```python
id2count = {}
for idd, iddx in enumerate(idx):
    id2count[iddx] = idd
idx_of_milano_in_order = { idd:k for k,idd in id2count.items()}
idx_in_order = [idd for k,idd in idx_of_milano_in_order.items()]    
id2count.get(2034)
```

```python
len(idx)
```

```python
# idx_of_milano_in_order
# idx_in_order
```

We generate the final distance matrix which shall be (31*39)^2 = 1209^2 = 1461681 ~ 1 462 000

```python
dist = np.zeros((len(idx_in_order),len(idx_in_order)), dtype = np.float32)
for row in dPow.itertuples():
#     print(row)
    dist[id2count.get(row[1]),id2count.get(row[2])]= row[3]
    dist[id2count.get(row[2]),id2count.get(row[1])]= row[3]
```

We transform the signal value so that 0 is mapped to 1 and increasing signal values make the distance converge to zero.
We use the function score : signal -> exp(-signal * alpha) where alpha is a parameter to determine

```python
exp_dist = dist
exp_dist[exp_dist==0] = np.nan
exp_dist = -np.log(exp_dist)
exp_dist = np.nan_to_num(exp_dist)
exp_dist[exp_dist==0.0] = 45
exp_dist= exp_dist-np.min(exp_dist)
pd_exp = pd.DataFrame(data = exp_dist, index = idx_in_order, columns = idx_in_order)
```

```python
# pd_exp.to_csv(fileout)
data = pd_exp
# fileout
dist[0,:],pd_exp.values[0,:]
```

#### Visualisation

```python
import matplotlib.pyplot as plt
import numpy as np
def plot_distance_mat(data, day = "2013-12-01", out = ""):
    fig = plt.figure(figsize=(12,6))
    plt.imshow(data, cmap='hot', interpolation='nearest')
    plt.title("Distance matrix - ")
#     plt.colorbar()  
    plt.show()
    if out:
        plt.savefig(out, bbox_inches='tight', pad_inches=0)
plot_distance_mat(data, out = "Results/"+file[:-4]+"-dist_mat-log.png")

```

```python
def permute_dist(data, idx,clusters,plot = True):
    o = list(zip(idx,clusters))
    o = list(sorted(o,key = lambda x : x[1]))
    perm = np.zeros_like(data)
    for i in range(data.shape[0]):
        perm[i] = np.copy(data[o[i][0]])
    perm2 = np.zeros_like(data)
    for j in range(data.shape[1]):
        perm[j] = np.copy(data[o[j][0]])
    if plot:
        plot_distance_mat(perm)
    return perm
```

```python
def plot_clusters(pd,labels = [], out = '', epsmin = None, k = None, clear = True):
    if epsmin == None:
        title = "Tripcolor of Milano"
    else :
        title = "Tripcolor of Milano - eps=" + str(epsmin[0])+" ; minPts="+str(epsmin[1])
    if k == None:
        title = "Tripcolor of Milano"
    else : 
        title = "Tripcolor of Milano - k=" + str(k)
        
    if len(labels):
        fig = plt.figure(figsize=(8,8))
        plt.title(title)
        plt.tripcolor(pd["i"], pd["j"], labels)
        plt.plot()
#         ax2.set_title("Tricontour of Milano")
#         ax2.tricontourf(pd["i"], pd["j"], labels)
    else:
        fig = plt.figure(figsize=(12,6))
        plt.title(title)
        plt.tripcolor(pd["i"], pd["j"], pd["Labels"])
        plt.plot()
    if out :
        plt.savefig(out, bbox_inches='tight', pad_inches=0)
    if clear : 
        plt.clf()
#         plt.title("Tricontour of Milano")
#         plt.tricontourf(pd["i"], pd["j"], pd["Labels"])
#         plt.plot()
#         fig, (ax, ax2)=plt.subplots(ncols=2)
#         ax.set_title("Tripcolor of Milano")
#         ax.tripcolor(pd["i"], pd["j"], pd["Labels"],edgecolors='k',shading='flat')
#         ax2.set_title("Tricontour of Milano")
#         ax2.tricontourf(pd["i"], pd["j"], pd["Labels"])
```

#### KMedoids
we use this implementation : https://github.com/letiantian/kmedoids


```python
from kmedoids import kMedoids
from sklearn.metrics.pairwise import pairwise_distances
import numpy as np
import pandas as pd
from matplotlib import gridspec
for k in range(1,10):
    print(k)
    M,C = kMedoids(np.array(data.values), k,tmax=300)
    Square = []
    Cluster = []
    for label in C:
        for point_idx in C[label]:
            Cluster.append(label)
            Square.append(idx_of_milano_in_order.get(point_idx))
    dSquare = pd.DataFrame(data=idx_in_order,columns=['Square'])
    dSquare['j'] = dSquare["Square"]%100
    dSquare['i'] = ((dSquare["Square"]-dSquare["j"])/100).apply(lambda x : int(x))
    dSquare["Labels"] = np.array(Cluster)
    dSquare.to_csv("Results/clusters_KMEDOIDS_k_"+str(k)+".csv",index=False)
    
    plot_clusters(dSquare, Cluster, out = "Results/clusters_KMEDOIDS_k_"+str(k)+".png", k = k, clear=False)
    #dist mat
    plt.figure()
    permuted = permute_dist(data.values, range(len(Cluster)),Cluster,plot = True)
    
```

```python
def plot_heatmap(Cluster):
    dSquare = pd.DataFrame(data=idx_in_order,columns=['Square'])
    dSquare['j'] = dSquare["Square"]%100
    dSquare['i'] = ((dSquare["Square"]-dSquare["j"])/100).apply(lambda x : int(x))
    dSquare["Labels"] = np.array(Cluster)
    M = dSquare.pivot("j","i","Labels")
    m = M.iloc[::-1]
    plt.figure(figsize=(11,12))
    plt.imshow(M.values, cmap='Set1', interpolation='nearest')
    plt.xticks(np.arange(len(M.columns)), M.columns)
    plt.yticks(sorted(np.arange(len(M.index)),reverse=True),M.index)
    plt.title("KMeans on distance matrix - k = "+str(k))
    plt.savefig("Results/clusters_KMEANS_k_"+str(k)+".png", bbox_inches='tight', pad_inches=0)
```

```python
from sklearn.cluster import KMeans
import seaborn as sns;
sns.set()
for k in np.arange(2, 10,step=1):
    KM = KMeans(n_clusters = k)
    Cluster = KM.fit_predict(data.values)
    plot_heatmap(Cluster)
M.values[15,15]
```

## Hierarchical clustering

```python
from scipy.cluster.hierarchy import single, dendrogram,set_link_color_palette, complete, average, centroid, ward,fcluster
```

```python
def plot_dendogramm(Z, out = '', title = ''):
    fig = plt.figure(figsize=(18, 10))
    dn1 = dendrogram(Z,orientation='top')
    # dn2 = dendrogram(Z, ax=axes[1],above_threshold_color='#bcbddc',orientation='right')
#     set_link_color_palette(None)  # reset to default after use
    if title:
        plt.title(title)
    plt.show()
    if out:
        plt.savefig(out, bbox_inches='tight', pad_inches=0 )
```

```python
Z = single(data.values)
plot_dendogramm(Z, out = "Results/Dendogram_SINGLE.png", title = "Dendrogram - Single linkage")
```

```python
Z = complete(data.values)
plot_dendogramm(Z, out = "Results/Dendogram_COMPLETE.png", title = "Dendrogram - Complete linkage")
```

```python
Z = average(data.values)
plot_dendogramm(Z, out = "Results/Dendogram_AVERAGE.png", title = "Dendrogram - AVERAGE linkage")
```

```python
Z = centroid(data.values)
plot_dendogramm(Z, out = "Results/Dendogram_CENTROID.png", title = "Dendrogram - CENTROID linkage")
```

```python
Z = ward(data.values)
plot_dendogramm(Z, out = "Results/Dendogram_WARD.png", title = "Dendrogram - WARD linkage")
```
